```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Total number of schools
total_schools = school_data_complete['school_name'].nunique()

# Total number of students
total_students = school_data_complete['Student ID'].nunique()

# Total budget (sum of unique budgets for each school)
total_budget = school_data_complete['budget'].unique().sum()

# Average maths score
average_maths_score = school_data_complete['maths_score'].mean()

# Average reading score
average_reading_score = school_data_complete['reading_score'].mean()

# Percentage of students with a passing maths score (50 or greater)
percent_passing_maths = (school_data_complete['maths_score'] >= 50).mean() * 100

# Percentage of students with a passing reading score (50 or greater)
percent_passing_reading = (school_data_complete['reading_score'] >= 50).mean() * 100

# Percentage of students who passed maths and reading (% Overall Passing)
percent_overall_passing = school_data_complete[(school_data_complete['maths_score'] >= 50) & (school_data_complete['reading_score'] >= 50)].shape[0] / total_students * 100

# Creating a dataframe to hold the results
summary_df = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Maths Score": [average_maths_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Maths": [percent_passing_maths],
    "% Passing Reading": [percent_passing_reading],
    "% Overall Passing": [percent_overall_passing]
})
# Formatting and displaying the data
summary_df['Total Students'] = summary_df['Total Students'].apply(lambda x: "{:,}".format(x))
summary_df['Total Budget'] = summary_df['Total Budget'].apply(lambda x: "${:,.2f}".format(x))
summary_df.head()
```

```python
# Trying to acheive stripes as in example notebook(return to this later)
if not isinstance(summary_df['Total Students'].iloc[0], str):
    summary_df['Total Students'] = summary_df['Total Students'].apply(lambda x: "{:,}".format(x))
if not isinstance(summary_df['Total Budget'].iloc[0], str):
    summary_df['Total Budget'] = summary_df['Total Budget'].apply(lambda x: "${:,.2f}".format(x))

# Correcting the styling to apply alternating grey and white stripes to rows instead of columns
styled_summary_df = summary_df.style.apply(lambda x: ["background: lightgrey" if i % 2 == 0 else "" for i in range(len(x))], axis=1)

styled_summary_df

```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Creating an overview table that summarises key metrics about each school

# Grouping the data by school
grouped_school_data = school_data_complete.groupby('school_name')

# Calculating the metrics for each school
school_summary = pd.DataFrame({
    "School Type": grouped_school_data['type'].first(),
    "Total Students": grouped_school_data.size(),
    "Total School Budget": grouped_school_data['budget'].first(),
    "Per Student Budget": grouped_school_data['budget'].first() / grouped_school_data.size(),
    "Average Maths Score": grouped_school_data['maths_score'].mean(),
    "Average Reading Score": grouped_school_data['reading_score'].mean(),
    "% Passing Maths": grouped_school_data.apply(lambda x: (x['maths_score'] >= 50).mean() * 100),
    "% Passing Reading": grouped_school_data.apply(lambda x: (x['reading_score'] >= 50).mean() * 100),
    "% Overall Passing": grouped_school_data.apply(lambda x: ((x['maths_score'] >= 50) & (x['reading_score'] >= 50)).mean() * 100)
})

school_summary.reset_index(inplace=True)
school_summary.rename(columns={'school_name': 'School Name'}, inplace=True)

school_summary.head(15)  # Display the first few rows of the dataframe

```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
# Sorting the school summary dataframe by '% Overall Passing' in descending order
# Displaying the top five performing schools
top_five_schools = school_summary.sort_values('% Overall Passing', ascending=False).head(5)
top_five_schools
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
# Sorting the school summary dataframe by '% Overall Passing' in ascending order
# Displaying the five worst-performing schools
bottom_five_schools = school_summary.sort_values('% Overall Passing', ascending=True).head(5)
bottom_five_schools
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Creating a table that lists the average maths score for students of each year level at each school

# Creating pandas series for each year
year_9_maths = school_data_complete[school_data_complete['year'] == 9].groupby('school_name')['maths_score'].mean()
year_10_maths = school_data_complete[school_data_complete['year'] == 10].groupby('school_name')['maths_score'].mean()
year_11_maths = school_data_complete[school_data_complete['year'] == 11].groupby('school_name')['maths_score'].mean()
year_12_maths = school_data_complete[school_data_complete['year'] == 12].groupby('school_name')['maths_score'].mean()

# Combining the series into a dataframe
maths_scores_by_year = pd.DataFrame({
    "9th": year_9_maths,
    "10th": year_10_maths,
    "11th": year_11_maths,
    "12th": year_12_maths
})

# Optional: cleaner formatting 
maths_scores_by_year = maths_scores_by_year

maths_scores_by_year.reset_index(inplace=True)
maths_scores_by_year.rename(columns={'school_name': 'School Name'}, inplace=True)

maths_scores_by_year  # Displaying the dataframe

```

## Reading Score by Year

* Perform the same operations as above for reading scores

```python
# Creating a table that lists the average reading score for students of each year level at each school

# Creating pandas series for each year
year_9_reading = school_data_complete[school_data_complete['year'] == 9].groupby('school_name')['reading_score'].mean()
year_10_reading = school_data_complete[school_data_complete['year'] == 10].groupby('school_name')['reading_score'].mean()
year_11_reading = school_data_complete[school_data_complete['year'] == 11].groupby('school_name')['reading_score'].mean()
year_12_reading = school_data_complete[school_data_complete['year'] == 12].groupby('school_name')['reading_score'].mean()

# Combining the series into a dataframe
reading_scores_by_year = pd.DataFrame({
    "9th": year_9_reading,
    "10th": year_10_reading,
    "11th": year_11_reading,
    "12th": year_12_reading
})

# Optional: cleaner formatting 
reading_scores_by_year = reading_scores_by_year

reading_scores_by_year.reset_index(inplace=True)
reading_scores_by_year.rename(columns={'school_name': 'School Name'}, inplace=True)

reading_scores_by_year  # Displaying the dataframe

```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Creating bins for school spending ranges
spending_bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]

# Categorizing the spending based on the bins
school_summary["Spending Ranges (Per Student)"] = pd.cut(school_summary["Per Student Budget"], spending_bins, labels=group_names, include_lowest=True)

# Grouping by spending ranges
grouped_by_spending = school_summary.groupby("Spending Ranges (Per Student)")

# Calculating averages and percentages
spending_summary = pd.DataFrame({
    "Average Maths Score": grouped_by_spending["Average Maths Score"].mean(),
    "Average Reading Score": grouped_by_spending["Average Reading Score"].mean(),
    "% Passing Maths": grouped_by_spending["% Passing Maths"].mean(),
    "% Passing Reading": grouped_by_spending["% Passing Reading"].mean(),
    "Overall Passing Rate": grouped_by_spending[["% Passing Maths", "% Passing Reading"]].mean().mean(axis=1)
})


spending_summary.reset_index(inplace=True)
spending_summary  # Display the table

```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Creating bins for school size ranges
size_bins = [0, 1000, 2000, 5000]
size_group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Categorizing the size based on the bins
school_summary["School Size"] = pd.cut(school_summary["Total Students"], size_bins, labels=size_group_names, include_lowest=True)

# Grouping by school size
grouped_by_size = school_summary.groupby("School Size")

# Calculating averages and percentages
size_summary = pd.DataFrame({
    "Average Maths Score": grouped_by_size["Average Maths Score"].mean(),
    "Average Reading Score": grouped_by_size["Average Reading Score"].mean(),
    "% Passing Maths": grouped_by_size["% Passing Maths"].mean(),
    "% Passing Reading": grouped_by_size["% Passing Reading"].mean(),
    "Overall Passing Rate": (grouped_by_size["% Passing Maths"].mean() + grouped_by_size["% Passing Reading"].mean()) / 2
})


size_summary.reset_index(inplace=True)
size_summary  # Display the table

```

```python
# Grouping by school type
grouped_by_type = school_summary.groupby("School Type")

# Calculating averages and percentages
type_summary = pd.DataFrame({
    "Average Maths Score": grouped_by_type["Average Maths Score"].mean(),
    "Average Reading Score": grouped_by_type["Average Reading Score"].mean(),
    "% Passing Maths": grouped_by_type["% Passing Maths"].mean(),
    "% Passing Reading": grouped_by_type["% Passing Reading"].mean(),
    "Overall Passing Rate": (grouped_by_type["% Passing Maths"].mean() + grouped_by_type["% Passing Reading"].mean()) / 2
})


type_summary.reset_index(inplace=True)
type_summary  # Display the table

```
